## import libraries

```python
%cd /content

from diffusers import StableDiffusionPipeline                                                                                                                                                                                                 

import torch
device = 'cuda' if torch.cuda.is_available() else "cpu"
print("using device is", device)
```

# load model

```python
model_id = "SG161222/Realistic_Vision_V1.4"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id)
pipe.to(device)
```

## Inference

```python
prompt = "attractive woman in business attire, light brown hair, portrait, modern, sleek, highly detailed, charming, pretty, glasses, perfect face features" # @param {type:"string"}
```

```python
image = pipe(
    prompt,
    generator = torch.Generator(device).manual_seed(12),
    num_inference_steps=52,
    width=512, height=720
    ).images[0]
```
